```python
#Importamos las dependencias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leemos el archivo
train = pd.read_csv('./titles.csv')
actors = pd.read_csv('./cast.csv')
#Rellenamos los campos null
train = train.fillna('')
#actors = train.fillna('')
```

```python
# Cual fue la primer pelicula hecha titulada "Romeo and Juliet" ?
train[train.title == "Romeo and Juliet"].sort_values('year').head(1)
```

```python
# Listar todas las peliculas que contengan la palabra "Exorcist" 
# ordenadas de la mas vieja a la mas reciente.
exorcist = train[train['title'].str.contains('Exorcist')]
exorcist.sort_values(by='year')
```

```python
# Listado de papeles de soporte que tuvo el actor "Charlton Heston" en
# la decada de los 50's, ordenado por a√±o
prot = actors[actors['year'].astype('int').between(1950,1959)]
prot1 = prot[prot['name'] == 'Charlton Heston']
prot1[prot1['n'] == 2]
```

```python
# Listado de papeles como protagonista que tuvo el actor "Charlton Heston" en
# la decada de los 60's, ordenado por a√±o de forma descendente
prot = actors[actors['year'].astype('int').between(1950,1959)]
prot1 = prot[prot['name'] == 'Charlton Heston']
prot1[prot1['n'] == 1]
```

```python
# ¬øCuantos papeles para actores hubo en la decada de los 50's?
temp = actors[actors['year'].astype('int').between(1950,1959)]
temp['type'][temp['type'] == 'actor'].count()
```
